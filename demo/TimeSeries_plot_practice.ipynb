# np.loadtxt 用法 :

1. **語法**
    - numpy.loadtxt(fname, comments='...', skiprows=..., usecols=..., unpack=...)

2. **變數**
    - fname : 檔案名稱
    - comments : 註解符號，預設為'#'。
    - skiprows : 跳過前面n行（預設為0）
    - usecols : 要讀入的直行數（預設為所有的直行都讀入）
    - unpack : 是否要將一個直行讀取成一個1D array，預設為關閉(False)，If set to 'True", x, y, z = loadtxt(...)
    
3. **回傳變數**
    - ndarray
    
[參考連結](https://docs.scipy.org/doc/numpy/reference/generated/numpy.loadtxt.html)

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('./201607_466920_hr.txt')
```

```python
P = data[:, 2]
T = data[:, 4]
Td = data[:, 6]
RH = data[:, 7]
WS = data[:, 9]
pp = data[:, 15]

## np.where用法 np.where(condition, A, B) 如果condition成立, 則為A, 否則為B
pp = np.where(pp > 0, pp, np.nan)
xx = np.arange(0, len(pp))
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#plt.style.use('classic')
```

```python
# split window
f, ax = plt.subplots(4,1,sharex='col', figsize=(15, 10))

ax[0].plot(T)
ax[0].plot(Td)
ax[0].set_ylabel('Temp')
ax[0].set_title('Time Series')
ax[0].set_xlim([150, 250])

ax2 = ax[0].twinx()
ax2.plot(RH, 'g', alpha=0.5)
ax2.set_ylabel('RH(%)')


ax[1].plot(P)
ax[1].set_ylabel('P(mb)')


ax[2].plot(WS)
ax[2].set_ylabel('Ws(m/s)')


ax[3].bar(xx, pp)
ax[3].set_ylabel('Rain(mm)')
ax[3].set_xlabel('Time')

#save figure and show
#plt.savefig('2x2_subplot.png')
plt.show()
```
